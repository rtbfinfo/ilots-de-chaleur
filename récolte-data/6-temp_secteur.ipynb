```python
import geopandas as gpd 
import pandas as pd 
```

```python
cities = ["charleroi", "liege", "mons", "namur", "tournai","bxl"]
every = pd.DataFrame()
for city in cities:
    df = gpd.read_file(f"../data/rev_fixed/{city}_sel_centroid.json")
    df2 = df.groupby("tx_sector_descr_fr")["raster_value"].median()
    test = pd.merge(df, df2, on='tx_sector_descr_fr')
    test.to_file(f"../data/rev_fixed/{city}_temp_fixed.json", driver="GeoJSON")
    test["city"] = city
    every = pd.concat([every, test], ignore_index=True)

every.to_file(f"../data/centroids/every_city.json", driver="GeoJSON")
every
```

```python
df = gpd.read_file(f"../assets/selection/bxl_sel_centroid.json")
print(df.shape)
df2 = df.groupby("tx_sector_descr_fr")["raster_value"].median()
test = pd.merge(df, df2, on='tx_sector_descr_fr')
test.to_file(f"../data/centroids/bxl_sel_centroid.json", driver="GeoJSON")
test.plot(aspect=1)
print(test.shape)
test
```
